```python
from graphviz import Digraph

# Data Class Flowchart
dot_data = Digraph(comment='Data Class Flowchart')
dot_data.attr(rankdir='TB')

# Title as an invisible node for spacing
dot_data.node('Title', 'Data Class', shape='plaintext')
dot_data.node('A', 'Load Dataset', shape='parallelogram')
dot_data.node('B', 'Preprocess Data', shape='box')
dot_data.node('C', 'Train-Test Split', shape='box')
dot_data.node('D', 'Balance Training Data\n(Only Training Set)', shape='box')

dot_data.edge('Title', 'A', style='invis')  # Invisible edge for space
dot_data.edges(['AB', 'BC', 'CD'])
dot_data.render('data_flowchart', format='png', view=True)


# QSVM Class Flowchart
dot_qsvm = Digraph(comment='QSVM Flowchart')
dot_qsvm.attr(rankdir='TB')

dot_qsvm.node('Title', 'QSVM Class', shape='plaintext')
dot_qsvm.node('E', 'Dimensionality Reduction (PCA)', shape='box')
dot_qsvm.node('F', 'Create Quantum Feature Map(ZZFeatureMap)', shape='box')
dot_qsvm.node('G', 'Intialize Quantum Kernel', shape='box')
dot_qsvm.node('H', 'Train Quantum Support Vector Machine', shape='box')
dot_qsvm.node('I', 'Evaluate Model Performance', shape='box')
dot_qsvm.node('J', 'Save Trained Model', shape='parallelogram')

dot_qsvm.edge('Title', 'E', style='invis')  # Invisible edge for space
dot_qsvm.edges(['EF', 'FG', 'GH', 'HI', 'HJ'])
dot_qsvm.render('qsvm_flowchart', format='png', view=True)


# Hybrid SVM Class Flowchart
dot_hybrid = Digraph(comment='Hybrid SVM Flowchart')
dot_hybrid.attr(rankdir='TB')

dot_hybrid.node('Title', 'Hybrid SVM Class', shape='plaintext')
dot_hybrid.node('E', 'Dimensionality Reduction (PCA)', shape='box')
dot_hybrid.node('F', 'Initialize Quantum Kernel', shape='box')
dot_hybrid.node('G', 'Hyperparameter Tuning (SVM)', shape='box')
dot_hybrid.node('H', 'Train Hybrid SVM', shape='box')
dot_hybrid.node('I', 'Evaluate Model Performance', shape='box')
dot_hybrid.node('J', 'Save Trained Model', shape='parallelogram')

dot_hybrid.edge('Title', 'E', style='invis')  # Invisible edge for space
dot_hybrid.edges(['EF', 'FG', 'GH', 'HI', 'HJ'])
dot_hybrid.render('hybrid_svm_flowchart', format='png', view=True)


# Classical SVM Class Flowchart
dot_svm = Digraph(comment='Classical SVM Flowchart')
dot_svm.attr(rankdir='TB')

dot_svm.node('Title', 'Classical SVM Class', shape='plaintext')
dot_svm.node('E', 'Apply PCA (Dimensionality Reduction)', shape='box')
dot_svm.node('F', 'Hyperparameter Tuning (Grid Search)', shape='box')
dot_svm.node('G', 'Train SVM Model', shape='box')
dot_svm.node('H', 'Evaluate Model', shape='box')
dot_svm.node('I', 'Save Trained Model', shape='parallelogram')

dot_svm.edge('Title', 'E', style='invis')  # Invisible edge for space
dot_svm.edges(['EF', 'FG', 'GH', 'GI'])
dot_svm.render('classical_svm_flowchart', format='png', view=True)
```
